```python
import pandas as pd 
```

```python
# Configure the path 
folder_path = r"F:\Megumi\DBOX-EQS1 Dropbox\Eqs1 Box03\Megumi-Cadisha\Skeleton by Lobe\Right Lower Lobe\proximal_pruning\13GWa-RL-P11\B9_B10\deloop5\deloop10\deloop20\\"

node_positions = pd.read_csv(folder_path + "node_positions.csv", delimiter = ",", names = ["Node_ID", "Y", "X", "Z"])
#node_positions.set_axis(["Node_ID", "Y", "X", "Z"], axis=1, inplace=True)
# Sanity check 
print(node_positions.head())
#edge_list_clean = pd.read_csv(folder_path + "edge_list_clean.csv", delimiter = ",")
# We create an empty .txt file to write to 
Node_Positions_txt = folder_path + "Node_Positions.txt"

# We will use this function to extract the node positions for a user-specified node 
def find_coordinates(node_positions, node):
    node_data = node_positions.loc[node_positions["Node_ID"] == int(node)]
    x, y, z = node_data.iloc[0][["X", "Y", "Z"]]
    return x, y, z
```

```python
# Input from user: 
# Specimen ID and Age 
ID_age = input("ID and Age: ")

# Root: 
root = input("Root: ")
#root=2618
# Find the corresponding coordinates 
root_x, root_y, root_z = find_coordinates(node_positions, root)
print("Root Node is", root, "XYZ is", "(", root_x, root_y, root_z, ")")

# B1and2: 
b6 = input("B6 (or NONE): ") 
if b6 == "NONE": 
    b6_x, b6_y, b6_z = ("nan", "nan", "nan") 
else: b6_x, b6_y, b6_z = find_coordinates(node_positions, b6)
print("B6 Node is", b6, "XYZ is", "(", b6_x, b6_y, b6_z, ")")

# B1and2: 
b7 = input("B7 (or NONE): ") 
if b7 == "NONE":
    b7_x, b7_y, b7_z = ("nan", "nan", "nan") 
else: b7_x, b7_y, b7_z = find_coordinates(node_positions, b7)
print("B7 Node is", b7, "XYZ is", "(", b7_x, b7_y, b7_z, ")")

# B3:
b8 = input("B8 (or NONE): ") 
if b8 == "NONE": 
    b8_x, b8_y, b8_z = ("nan", "nan", "nan") 
else: b8_x, b8_y, b8_z = find_coordinates(node_positions, b8)
print("B8 Node is", b8, "XYZ is", "(", b8_x, b8_y, b8_z, ")")

# B4:
b9 = input("B9 (or NONE): ") 
if b9 == "NONE":
    b9_x, b9_y, b9_z = ("nan", "nan", "nan") 
else: b9_x, b9_y, b9_z = find_coordinates(node_positions, b9)
print("B9 Node is", b9, "XYZ is", "(", b9_x, b9_y, b9_z, ")")

# B5: 
b10 = input("B10: ") 
if b10 == "NONE": 
    b10_x, b10_y, b10_z = ("nan", "nan", "nan") 
else: b10_x, b10_y, b10_z = find_coordinates(node_positions, b10)
print("B10 Node is", b10, "XYZ is", "(", b10_x, b10_y, b10_z, ")")

# Additional B* Branch
bsharp = input("B Sharp (or NONE): ")
if bsharp == "NONE":
    bsharp_x, bsharp_y, bsharp_z = ("nan", "nan", "nan") 
else: bsharp_x, bsharp_y, bsharp_z = find_coordinates(node_positions, bsharp) 
print("Bsharp Node is", bsharp, "XYZ is", "(", bsharp_x, bsharp_y, bsharp_z, ")")

# Additional comments 
comments = input("Comments: ") 


# Write a text file with all of the information
with open(Node_Positions_txt, "w") as file: 
    file.write(str(ID_age))
    file.write("\n" + "Node positions in (Y, X, Z)")
    file.write("\n" + "Root: Node " + str(root) + " (" + str(root_y) + ", " + str(root_x) + ", " + str(root_z) + ") ")
    file.write("\n" + "B6: Node " + str(b6) + " (" + str(b6_y) + ", " + str(b6_x) + ", " + str(b6_z) + ") ")
    file.write("\n" + "B7: Node " + str(b7) + " (" + str(b7_y) + ", " + str(b7_x) + ", " + str(b7_z) + ") ")
    file.write("\n" + "B8: Node " + str(b8) + " (" + str(b8_y) + ", " + str(b8_x) + ", " + str(b8_z) + ") ")
    file.write("\n" + "B9: Node " + str(b9) + " (" + str(b9_y) + ", " + str(b9_x) + ", " + str(b9_z) + ") ")
    file.write("\n" + "B10: Node " + str(b10) + " (" + str(b10_y) + ", " + str(b10_x) + ", " + str(b10_z) + ") ")
    file.write("\n" + "Bsharp: Node " + str(bsharp) + " (" + str(bsharp_y) + ", " + str(bsharp_x) + ", " + str(bsharp_z) + ") ")
    #file.write("\n" + "Total_Clean_Edges: " + str(edge_clean_num))
    file.write("\n" + "Comments: " + str(comments))
    file.write("\n" + str(root_y) + " " + str(root_x) + " " + str(root_z) + ";" + str(b6_y) + " " + str(b6_x) + " " + str(b6_z) + ";" +
               str(b7_y) + " " + str(b7_x) + " " + str(b7_z) + ";" +
               str(b8_y) + " " + str(b8_x) + " " + str(b8_z) + ";" +
               str(b9_y) + " " + str(b9_x) + " " + str(b9_z) + ";" +
               str(b10_y) + " " + str(b10_x) + " " + str(b10_z) + ";" +
               str(bsharp_y) + " " + str(bsharp_x) + " " + str(bsharp_z))

print("Text file is saved!")
```
