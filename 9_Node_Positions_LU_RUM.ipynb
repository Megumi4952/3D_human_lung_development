```python
import pandas as pd 
```

```python
# Configure the path 
folder_path = r"F:\Megumi\Dropbox (DBOX-EQS1)\Megumi-Cadisha\Skeleton by Lobe\Left Upper Lobe\proximal_pruning\EH4335LU-P9.3(9.7)\\"
node_positions = pd.read_csv(folder_path + "node_positions.csv", delimiter = ",", names=["Node_ID", "Y", "X", "Z"])
#node_positions.set_axis(["Node_ID", "Y", "X", "Z"], axis=1, inplace=True)
# Sanity check 
print(node_positions.head())
#edge_list_clean = pd.read_csv(folder_path + "edge_list_clean.csv", delimiter = ",")
# We create an empty .txt file to write to 
Node_Positions_txt = folder_path + "Node_Positions.txt"

# We will use this function to extract the node positions for a user-specified node 
def find_coordinates(node_positions, node):
    node_data = node_positions.loc[node_positions["Node_ID"] == int(node)]
    x, y, z = node_data.iloc[0][["X", "Y", "Z"]]
    return x, y, z
```

```python
###### LEFT UPPER 
# Input from user: 
# Specimen ID and Age 
ID_age = input("ID and Age: ")

# Root: 
root = input("Root: ")
#root=2618
# Find the corresponding coordinates 
root_x, root_y, root_z = find_coordinates(node_positions, root)
print("Root Node is", root, "XYZ is", "(", root_x, root_y, root_z, ")")


# B1and2: 
b1and2 = input("B1and2  (or NONE): ") 
if b1and2 == "NONE":
    b1and2_x, b1and2_y, b1and2_z = ("nan", "nan", "nan") 
else: b1and2_x, b1and2_y, b1and2_z = find_coordinates(node_positions, b1and2) 
print("B1and2 Node is", b1and2, "XYZ is", "(", b1and2_x, b1and2_y, b1and2_z, ")")

# B3:
b3 = input("B3 (or NONE): ") 
if b3 == "NONE": 
    b3_x, b3_y, b3_z = ("nan", "nan", "nan") 
else: b3_x, b3_y, b3_z = find_coordinates(node_positions, b3)
print("B3 Node is", b3, "XYZ is", "(", b3_x, b3_y, b3_z, ")")

# B4:
b4 = input("B4 (or NONE): ") 
if b4 == "NONE": 
    b4_x, b4_y, b4_z =  ("nan", "nan", "nan") 
else: b4_x, b4_y, b4_z = find_coordinates(node_positions, b4)
print("B4 Node is", b4, "XYZ is", "(", b4_x, b4_y, b4_z, ")")

# B5: 
b5 = input("B5 (or NONE): ") 
if b5 == "NONE": 
    b5_x, b5_y, b5_z =  ("nan", "nan", "nan") 
else: b5_x, b5_y, b5_z = find_coordinates(node_positions, b5)
print("B5 Node is", b5, "XYZ is", "(", b5_x, b5_y, b5_z, ")")

# Additional comments 
comments = input("Comments: ") 


# Write a text file with all of the information
with open(Node_Positions_txt, "w") as file: 
    file.write(str(ID_age))
    file.write("\n" + "Node positions in (Y, X, Z)")
    file.write("\n" + "Root: Node " + str(root) + " (" + str(root_y) + ", " + str(root_x) + ", " + str(root_z) + ") ")
    file.write("\n" + "B1and2: Node " + str(b1and2) + " (" + str(b1and2_y) + ", " + str(b1and2_x) + ", " + str(b1and2_z) + ") ")
    file.write("\n" + "B3: Node " + str(b3) + " (" + str(b3_y) + ", " + str(b3_x) + ", " + str(b3_z) + ") ")
    file.write("\n" + "B4: Node " + str(b4) + " (" + str(b4_y) + ", " + str(b4_x) + ", " + str(b4_z) + ") ")
    file.write("\n" + "B5: Node " + str(b5) + " (" + str(b5_y) + ", " + str(b5_x) + ", " + str(b5_z) + ") ")
    #file.write("\n" + "Total_Clean_Edges: " + str(edge_clean_num))
    file.write("\n" + "Comments: " + str(comments))
    file.write("\n" + str(root_y) + " " + str(root_x) + " " + str(root_z) + ";" + str(b1and2_y) + " " + str(b1and2_x) + " " + str(b1and2_z) + ";" +
               str(b3_y) + " " + str(b3_x) + " " + str(b3_z) + ";" +
               str(b4_y) + " " + str(b4_x) + " " + str(b4_z) + ";" +
               str(b5_y) + " " + str(b5_x) + " " + str(b5_z))

print("Text file is saved!")
```
